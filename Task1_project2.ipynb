```python
import pandas as pd
import datetime as dt
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('/home/jupyter-d.fomicheva-1/Untitled Folder/test1_completed.csv')
```

```python
#checking if quantity isnt zero
data=data[data['Количество']>0]
```

```python
#building a matrix based on id
basket=data.groupby(['id','Товар'])['Количество'].sum().unstack().reset_index().fillna(0).set_index('id')
```

```python
#making the set binary
def encode_units(x):
    if x<=0:
        return 0
    if x>=1:
        return 1
    
    
basket_new=basket.applymap(encode_units)
```

```python
#get rid of NAs
basket_new=basket_new.dropna()
```

```python
#at least 2 items are bought
basket_filt=basket_new[(basket_new>0).sum(axis=1)>=2]
```

```python
basket_filt
```

```python
pip install mlxtend  
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
#apply apriori. lets consider a frequently bought item as a 3% chance
basket_apr = apriori(basket_filt, min_support=0.03,
                     use_colnames=True).sort_values('support', ascending=False).reset_index(drop=True)

basket_apr['length']=basket_apr['sets'].apply(lambda x: len(x))
```

```python
#lets find out what items are most frequently bought together, so we could recommend them
association_rules(basket_apr, metric='lift',
                  min_threshold=1).sort_values('lift', ascending=False).reset_index(drop=True)
```
